```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('/content/drive/MyDrive/Customers.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Products.csv')
df3 = pd.read_csv('/content/drive/MyDrive/Transactions.csv')
```

```python
print("\nDataset 1 Statistics:")
print(df1.describe())

print("\nDataset 2 Statistics:")
print(df2.describe())

print("\nDataset 3 Statistics:")
print(df3.describe())
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df1.select_dtypes(include=[np.number]).melt(value_name='value').dropna(), kde=True)
plt.title('Distribution of Numerical Features in Dataset 1')
plt.show()

plt.figure(figsize=(12, 6))
sns.histplot(df2.select_dtypes(include=[np.number]).melt(value_name='value').dropna(), kde=True)
plt.title('Distribution of Numerical Features in Dataset 2')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df3.select_dtypes(include=[np.number]).melt(value_name='value').dropna(), kde=True)
plt.title('Distribution of Numerical Features in Dataset 3')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
# Select only numerical columns for correlation calculation
numerical_df1 = df1.select_dtypes(include=[np.number])

# Check if numerical_df1 is empty or has only NaNs
if numerical_df1.empty or numerical_df1.isnull().all().all():
    print("No numerical data available for correlation calculation.")
else:
    sns.heatmap(numerical_df1.corr(), annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Heatmap for Dataset 1')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
# Select only numerical columns for correlation calculation
numerical_df2 = df2.select_dtypes(include=[np.number])
```

```python
plt.figure(figsize=(10, 6))
# Select only numerical columns for correlation calculation
numerical_df3 = df3.select_dtypes(include=[np.number])
```

```python
# Check if the DataFrame has any numerical columns after dropping NaNs
numerical_df1 = df1.select_dtypes(include=[np.number]).dropna()

if numerical_df1.empty:
    print("No numerical data available for pairplot for Dataset 1. Skipping...")
else:
    sns.pairplot(numerical_df1)
    plt.title('Pairplot for Dataset 1')
    plt.show()

# Repeat the same check for df2 and df3
numerical_df2 = df2.select_dtypes(include=[np.number]).dropna()

if numerical_df2.empty:
    print("No numerical data available for pairplot for Dataset 2. Skipping...")
else:
    sns.pairplot(numerical_df2)
    plt.title('Pairplot for Dataset 2')
    plt.show()

numerical_df3 = df3.select_dtypes(include=[np.number]).dropna()

if numerical_df3.empty:
    print("No numerical data available for pairplot for Dataset 3. Skipping...")
else:
    sns.pairplot(numerical_df3)
    plt.title('Pairplot for Dataset 3')
    plt.show()
```

```python
cat_columns_df1 = df1.select_dtypes(include=['object']).columns
for col in cat_columns_df1:
    plt.figure(figsize=(8, 4))
    sns.countplot(y=df1[col])
    plt.title(f'Countplot of {col} in Dataset 1')
    plt.show()
```
